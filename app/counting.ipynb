```python
from scipy.ndimage import measurements, morphology
from matplotlib import pylab as plb
from PIL import Image
import cv2
import numpy as np
```

```python
def show_img(img):
    plb.gray()
    plb.figure(figsize=(18,18))
    plb.imshow(img)
    plb.show()
```

```python
img = np.array(Image.open("../sac.png").convert("L"))
show_img(img)
```

```python
img = 1*(img<128)
show_img(img)
```

```python
labled_array, count_labels = measurements.label(img)
labled_array, count_labels
```

```python
show_img(labled_array)
```

```python
img = np.log2(255 - np.array(Image.open("../sac.png").convert("L")))
show_img(img*20)
```

```python
img = 1*((img*10)>70)
show_img(img)
```

```python
bin_img = morphology.binary_opening(img, np.ones((9,5)), iterations=2)
show_img(bin_img)
```

```python
labled_array, count_labels = measurements.label(img)
labled_array, count_labels
```

```python
show_img(255-labled_array)
```

```python
plb.imsave("bin_img.jpg", bin_img)
```
